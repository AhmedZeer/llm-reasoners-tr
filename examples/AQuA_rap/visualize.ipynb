```python
import re

def retrieve_answer(output):
    '''
    output should be a world_model.MATHState if being a list
    '''
    print('retrieve_answer:', output)
    
    answers = []  
    for output_i in output:
        match = re.match(r'.*[Tt]he answer is.*?([A-E]).*?$', output_i, re.DOTALL)
        
        if match is None:
            print('match:', match)
            return None
        answer = match[1].strip()
        answers.append(answer)
    counter = Counter(answers)
    answer, _ = counter.most_common(1)[0]
    print(f'retrived answer: {answer}, {answers}')
    return answer
```
